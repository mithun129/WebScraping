### 7. Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.amazon.in/Mobile-Phone-Under-20000-Rupees/s?k=Mobile+Phone+Under+20000+Rupees")
```

```python
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content, 'html.parser')
```

```python
soup
```

```python
name=soup.find_all('span',class_="a-size-medium a-color-base a-text-normal")
```

```python
name
```

```python
names=[]
for i in range(0,len(name)):
    names.append(name[i].get_text())
```

```python
names
```

```python
price=soup.find_all('span',class_="a-price-whole")
```

```python
price
```

```python
prices=[]
for i in range(0,len(price)):
    prices.append(price[i].get_text())
```

```python
prices
```

```python
rating=soup.find_all('i',class_="a-icon a-icon-star-small a-star-small-4 aok-align-bottom")
```

```python
rating
```

```python
ratings=[]
for i in range(0,len(rating)):
    ratings.append(rating[i].get_text())
```

```python
ratings
```
