#### 4. write a python program to scrap the book name, author_name, genre and book review

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://bookpage.com/reviews")
```

```python
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content, 'html.parser')
```

```python
soup
```

```python
book_name=soup.find_all('h4',class_="italic")
```

```python
book_name
```

```python
author_name=soup.find_all('p', class_="sans bold")
```

```python
author_name
```

```python
review=soup.find_all('p',class_="excerpt")
```

```python
review
```

```python
genre=soup.find_all('p',class_="genre-links hidden-phone")
```

```python
genre
```

```python
import pandas as pd
Book=pd.DataFrame({})
Book['Book_name']=book_name[0:14]
Book['Author_name']=author_name[0:14]
Book['Review']=review[0:14]
Book['Genre']=genre[0:14]
Book
```
