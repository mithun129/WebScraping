##### ii) top 10 ODI Batsmen in men along with record of their team and rating

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/test/batting")
```

```python
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content, 'html.parser')
```

```python
soup
```

```python
Batsmen=soup.find_all('td',class_="table-body__cell rankings-table__name name")                        
```

```python
Batsmen
```

```python
team=soup.find_all('td',class_="table-body__cell nationality-logo rankings-table__team")                          
```

```python
team
```

```python
rating=soup.find_all('td',class_="table-body__cell rating")       
```

```python
rating
```

```python
import pandas as pd
cricket=pd.DataFrame({})
cricket['batsmen']=Batsmen[0:14]
cricket['teams']=team[0:14]
cricket['ratings']=rating[0:14]
cricket
```
