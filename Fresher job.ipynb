### 9. Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, company name, CTC, and apply date.

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://internshala.com/fresher-jobs")
```

```python
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content, 'html.parser')
```

```python
soup
```

```python
job_title=soup.find_all('div',class_="heading_4_5 profile")
```

```python
job_title
```

```python
company_name =soup.find_all('div', class_="heading_6 company_name")
```

```python
company_name
```

```python
CTC=soup.find_all('div',class_="other_detail_item")
```

```python
CTC
```

```python
apply_date=soup.find_all('div',class_="other_detail_item_row")
# apply_date = soup.div.find('div', class_ = "item_body").text.replace('(', '').replace(')', '')
# soup.append(apply_date)
```

```python
apply_date
```

```python
import pandas as pd
fresher_job=pd.DataFrame({})
fresher_job['Job_Title']=job_title[0:12]
fresher_job['Company_Name']=company_name[0:12]
fresher_job['CTC']=CTC[0:12]
fresher_job['Apply_Date']=apply_date[0:12]
fresher_job
```
