### 1. Write a python program to display all the header tags from
‘en.wikipedia.org/wiki/Main_Page’.
#### First, let's import the required modules

```python
# importing required libraries
import requests
from bs4 import BeautifulSoup   # We import the BeautiFulsoup Class.
from urllib.request import urlopen
```

###### Now, We load the htmlsource directly from a website using the requests moddule:

```python
header = requests.get('https://en.wikipedia.org/wiki/Main_Page')
header
```

.Lets see show the webpage source code downloaded above looks like

```python
header.content
```

```python
soup = BeautifulSoup(header.content)
soup
```

```python
print(soup.prettify())
```

###### Extract all the header tags from ‘en.wikipedia.org/wiki/Main_Page’

```python
html = urlopen('https://en.wikipedia.org/wiki/Main_Page')
bs = BeautifulSoup(html, "html.parser")
titles = bs.find_all(['h1', 'h2','h3','h4','h5','h6']) 
print('List all the header tags :', *titles, sep='\n\n')
```
